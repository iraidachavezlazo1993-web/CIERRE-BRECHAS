```python
# Corrección de codigos locales
def codlocal(df, oldvar, newvar):
    df[newvar] = df[oldvar].astype(str).str.zfill(6)
```

## 02. Identificación de LL.EE. publicos existentes en 2013

```python
file = 'data/Padrones/Padron activos 2013.xlsx'
df_0 = pd.read_excel(file)
```

```python
df = df_0.copy()

# Corrigiendo codigos locales
codlocal(df, 'CODLOCAL', 'codlocal')

# Identificando LL.EE. activos
print(df['ESTADO'].value_counts(dropna=False), '\n')

# Identificando LL.EE. bajo gestion publica
print(df['GESTION'].value_counts(dropna=False), '\n')
df = df[df['GESTION'].ne('Privada')]

# Identificando LL.EE. escolarizados
print(df['NIVEL/MODALIDAD/ETAPA'].value_counts(dropna=False), '\n')
df = df[df['NIVEL/MODALIDAD/ETAPA'].ne('Inicial - Programa no escolarizado')]
df['codlocal_len'] = df['codlocal'].astype('string').str.len()
print(df['codlocal_len'].value_counts(dropna=False))
```

```python
# Seleccionando variables
df = df['codlocal']
df.drop_duplicates(inplace=True)
df.to_csv('data/procesadas/llee_existentes.csv', index=False)
```
